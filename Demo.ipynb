```python
import pandas as pd
```

```python
# Read in data
data = pd.read_table('data/amazon_cells_labelled.txt', header = None)
data.columns = ['Text', 'Label']
```

```python
data.head()
```

```python
def load_data_into_text(filename, text_file, tag):
    data = pd.read_table(text_file, names = ['Text', 'Label'])
    with open(filename, "w") as f:
        f.writelines(["%s\n" % i for i in data[data.Label==tag].Text])
    f.close()

```

```python
load_data_into_text("positive.txt", "amazon_cells_labelled.txt", 1)
```

```python
load_data_into_text("negative.txt", "amazon_cells_labelled.txt", 0)
```
